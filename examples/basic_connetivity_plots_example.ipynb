# Basic connectivity plots

Example notebook on how to generate basic connectivity plots from a ConnectivityMatrix object

```python
import obi
```

```python
basic_connectivity_plots_form = obi.BasicConnectivityPlots(
                    initialize=obi.BasicConnectivityPlots.Initialize(
                        matrix_path=[obi.NamedPath(name="Human_L23", path="/Users/danielaegas/Desktop/Human_pruned_E_to_E_conmat.h5"),
                                     obi.NamedPath(name="Rat_2015", path="/Users/danielaegas/Desktop/connectome.h5")],
                        plot_formats=("png",)
                        )
                        )

grid_scan = obi.GridScan(form=basic_connectivity_plots_form, output_root='../../obi-output/basic_connectivity_plots/grid_scan', coordinate_directory_option="VALUE")
```

```python
grid_scan.run()
```

```python
# Display the intermediary data
grid_scan.multiple_value_parameters(display=True)
grid_scan.coordinate_parameters(display=True)
grid_scan.coordinate_instances(display=True)
```

```python
# Deserialize
grid_scan_ds = obi.deserialize_obi_object_from_json_file("../../obi-output/basic_connectivity_plots/grid_scan/run_scan_config.json")
grid_scan_ds 
```
