# Generate a Toy SpikeInterface Recording

Installs `spikeinterface` into the current venv and generates a small ground-truth recording that can be used as input to the AIND ephys dispatch pipeline. The recording is saved to `obi-output/00_toy_example_recording/` (the same `obi-output/` tree every downstream notebook writes to).

## Install spikeinterface

Uses `uv pip install` so the package is added to the active project venv.

```python
import sys
import subprocess

subprocess.run(["uv", "pip", "install", "--python", sys.executable, "spikeinterface"], check=True)
```

## Generate and save a ground-truth recording

```python
from pathlib import Path

import spikeinterface.full as si

output_dir = (Path.cwd() / "../../../obi-output/00_toy_example_recording").resolve()
output_dir.parent.mkdir(parents=True, exist_ok=True)

durations = [10.0]
job_kwargs = {"n_jobs": 1, "chunk_duration": "1s"}

rec, _ = si.generate_ground_truth_recording(
    durations=durations,
    sampling_frequency=30_000.0,
    num_channels=70,
    num_units=10,
    seed=2205,
)
rec.save(folder=str(output_dir), verbose=False, **job_kwargs)

recording = si.load(str(output_dir))
print("Saved to:", output_dir)
recording
```
