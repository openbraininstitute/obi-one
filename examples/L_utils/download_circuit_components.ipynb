# Download circuit components from entitycore

This notebook demonstrates how to download individual components of a circuit:
- Circuit config
- Node sets
- Electrical models (hoc files)
- Mechanisms (mod files)
- Node populations (h5 files)
- Edge populations (h5 files)
- ID mapping

## Setup

```python
from pathlib import Path

from entitysdk import Client, ProjectContext
from obi_auth import get_token
```

```python
token = get_token(environment="staging")
project_context = ProjectContext(
    virtual_lab_id="e6030ed8-a589-4be2-80a6-f975406eb1f6",
    project_id="2720f785-a3a2-4472-969d-19a53891c817",
)
client = Client(environment="staging", project_context=project_context, token_manager=token)
```

## Circuit ID and output directory

```python
circuit_id = "a2518876-eb72-4a1e-8888-b5051ddaf3b2"
dest_dir = Path("downloaded_components")
```

## Download circuit config

```python
from obi_one.utils.circuit_customization.download import download_circuit_config

config_path = download_circuit_config(circuit_id, client, dest_dir / "config")
print(f"Circuit config downloaded to: {config_path}")
```

## Download node sets

```python
from obi_one.utils.circuit_customization.download import download_node_sets

node_sets_path = download_node_sets(circuit_id, client, dest_dir / "node_sets")
print(f"Node sets downloaded to: {node_sets_path}")
```

## Download ID mapping

```python
from obi_one.utils.circuit_customization.download import download_id_mapping

id_mapping_path = download_id_mapping(circuit_id, client, dest_dir / "id_mapping")
print(f"ID mapping downloaded to: {id_mapping_path}")
```

## Download electrical models (hoc files)

Downloads hoc files for all node populations (into subfolders by population name).
Optionally, specify `node_population` to download for a single population only.

```python
from obi_one.utils.circuit_customization.download import download_electrical_models

hoc_files = download_electrical_models(circuit_id, client, dest_dir / "electrical_models")
print(f"Downloaded {len(hoc_files)} hoc file(s)")
```

## Download mechanisms (mod files)

```python
from obi_one.utils.circuit_customization.download import download_mechanisms

mod_files = download_mechanisms(circuit_id, client, dest_dir / "mechanisms")
print(f"Downloaded {len(mod_files)} mod file(s)")
```

## Download node populations (h5 files)

Downloads node population files for all populations (into subfolders by population name).
Optionally, specify `node_population` to download for a single population only.

```python
from obi_one.utils.circuit_customization.download import download_node_populations

node_files = download_node_populations(circuit_id, client, dest_dir / "node_populations", "hippocampus_neurons")
print(f"Downloaded {len(node_files)} node population file(s)")
```

## Download edge populations (h5 files)

Downloads edge population files for all populations (into subfolders by population name).
Optionally, specify `edge_population` to download for a single population only.

```python
from obi_one.utils.circuit_customization.download import download_edge_populations

edge_files = download_edge_populations(circuit_id, client, dest_dir / "edge_populations", "hippocampus_neurons__hippocampus_neurons__chemical_synapse")
print(f"Downloaded {len(edge_files)} edge population file(s)")
```
