# Basic connectivity plots

Example notebook on how to generate basic connectivity plots from a ConnectivityMatrix object

```python
import obi_one as obi
from pathlib import Path
```

```python
root = "/Users/danielaegas/OneDrive - Open Brain Institute/Shared Documents - OBI - Scientific staff/Data/Circuits hardcoded/test_tiny_circuits/ConnectivityMatrices/"
mat_path1=root+"N_10__selection1/connectivity_matrix.h5"
mat_path2=root+"N_10__selection2/connectivity_matrix.h5"

basic_connectivity_plots_form = obi.BasicConnectivityPlots(initialize=obi.BasicConnectivityPlots.Initialize(
    matrix_path=[obi.NamedPath(name="Small_test1", path=mat_path1),
                 obi.NamedPath(name="Small_test2", path=mat_path2)],
                 plot_formats=("png",),  # sub-tuple of ('png', 'pdf', a'svg'), if not specified all are plotted
                 #plot_types=("nodes",), # sub-tuple of ('nodes', 'connectivity_global', 'connectivity_pathway'), if not specified all are plotted
                 # # dpi=int default 300
                 )
                 )

grid_scan = obi.GridScan(form=basic_connectivity_plots_form, output_root="../../obi-output/basic_connectivity_plots/grid_scan", coordinate_directory_option="VALUE")
```

```python
grid_scan.execute(processing_method="run")
```

```python
# Display the intermediary data
grid_scan.multiple_value_parameters(display=True)
grid_scan.coordinate_parameters(display=True)
grid_scan.coordinate_instances(display=True)
```
